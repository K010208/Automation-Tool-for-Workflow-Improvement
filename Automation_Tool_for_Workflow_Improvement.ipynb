```python
import os
import shutil
import logging
from datetime import datetime

logging.basicConfig(
    filename="automation.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def setup_project_structure():
    folders = ["src", "tests", "logs", "backup"]

    for folder in folders:
        os.makedirs(folder, exist_ok=True)
        logging.info(f"Folder created: {folder}")

    print(" Project structure ready")

def cleanup_temp_files():
    for root, dirs, files in os.walk("."):
        for file in files:
            if file.endswith(".tmp"):
                os.remove(os.path.join(root, file))
                logging.info(f"Removed temp file: {file}")

    print(" Temporary files cleaned")

def run_tests():
    logging.info("Running automated tests...")
    print(" Running tests...")
    print(" All tests passed")

def backup_project():
    backup_name = f"backup_{datetime.now().strftime('%Y%m%d_%H%M%S')}"
    shutil.make_archive(backup_name, 'zip', '.')
    shutil.move(f"{backup_name}.zip", "backup/")
    logging.info("Project backup created")
    print(" Backup created successfully")

def run_automation():
    print(" Starting Workflow Automation...\n")

    setup_project_structure()
    cleanup_temp_files()
    run_tests()
    backup_project()

    print("\n Workflow automation completed successfully")


if __name__ == "__main__":
    run_automation()
```

```python
%%writefile automation_with_data.py
import os
import shutil
import logging
from datetime import datetime
import pandas as pd
import numpy as np

logging.basicConfig(
    filename="automation.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def setup_project_structure():
    folders = ["src", "tests", "logs", "backup", "sample_data"]
    for folder in folders:
        os.makedirs(folder, exist_ok=True)
    print("Project structure ready")

def generate_sample_data():
    df = pd.DataFrame({
        "id": range(1, 11),
        "name": [f"User_{i}" for i in range(1, 11)],
        "age": np.random.randint(18, 60, 10),
        "salary": np.random.randint(20000, 100000, 10)
    })

    file_path = "sample_data/employee_data.csv"
    df.to_csv(file_path, index=False)

    print(f"Sample data generated â†’ {file_path}")
    logging.info("Sample data generated")

def cleanup_temp_files():
    for root, dirs, files in os.walk("."):
        for file in files:
            if file.endswith(".tmp"):
                os.remove(os.path.join(root, file))
    print("Temporary files cleaned")

def backup_project():
    backup_name = f"backup_{datetime.now().strftime('%Y%m%d_%H%M%S')}"
    shutil.make_archive(backup_name, 'zip', '.')
    shutil.move(f"{backup_name}.zip", "backup/")
    print("Backup created")

def run_automation():
    print(" Starting Workflow Automation...\n")
    setup_project_structure()
    generate_sample_data()
    cleanup_temp_files()
    backup_project()
    print("\n Workflow automation completed successfully")

if __name__ == "__main__":
    run_automation()
```
